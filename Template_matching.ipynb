```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_rgb = cv2.imread('C://Users//lenovo//Desktop//Python stuff//IP_with_python//actual_image_1.png')
#converting the target image to gray
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY) 

template = cv2.imread('C://Users//lenovo//Desktop//Python stuff//IP_with_python//tofind_1.png',0)
template_1 = cv2.imread('C://Users//lenovo//Desktop//Python stuff//IP_with_python//tofind_1.png',1)
plt.imshow(template_1)
plt.title('Template')
plt.show()

#T0 know the dimensions of the template
w, h = template.shape[: : -1]

#Using the template matching function of opencv 
#and setting a threshold for image detection
res = cv2.matchTemplate(img_gray,template,cv2.TM_CCOEFF_NORMED)
threshold = 0.8
loc = np.where( res >= threshold)

#making the box on original image showing the matched template
for pt in zip(*loc[: : -1]):
    cv2.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0,255,255), 2)

#cv2.imwrite('Detected',img_rgb)
plt.imshow(img_rgb)
plt.title('Detected')
plt.show()
```
